# Import Library

```python
import numpy as np
import pandas as pd
import csv
import re
```

# Functions

# Read CSV

```python
df_book = pd.read_csv("data/books_db.csv", encoding="utf-8")
book = df_book[
    [
        "book_title",
        "price",
        "images",
        "desc",
        "author",
        "pushlisher",
        "publication_date",
        "page",
        "product_dimensions",
        "rating",
    ]
]
images = [im.replace("'", "") for im in book["images"]]
prices = [p.replace("$", "") for p in book["price"]]
# desc = [d.replace('"', '\\"').replace("'", "\\'").replace('“','\\"') for d in book["desc"]]
book["images"] = images
book["price"] = prices
# book["desc"] = desc
book.insert(loc=9, column="quantity", value=[1000 for i in range(len(book.values))])
book.rename(
    columns={
        "book_title": "name",
        "desc": "description",
        "images": "image",
        "pushlisher": "publisher",
    },
    inplace=True,
)
book
```

# Data

```python
columns = [
    "name",
    "price",
    "image",
    "description",
    "author",
    "publisher",
    "publication_date",
    "page",
    "product_dimensions",
    "quantity",
    "rating",
]
df_books = pd.DataFrame(data=book, columns=columns)
df_books
```

```python
df_books.to_csv(
    "data/books.csv", sep=",", index=False, encoding="utf-8", index_label=None
)
```
