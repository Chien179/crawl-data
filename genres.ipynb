# Functions

# Read CSV

```python
df_book = pd.read_csv("data/books_db.csv", encoding="utf-8")
cate = [
    c.strip()
    for ct in df_book["category"]
    for c in ct.replace("[", "")
    .replace("]", "")
    .replace("'", "")
    .replace("&amp;", "&")
    .replace("&gt;", ">")
    .split(",")
]
```

```python
genres = []
subgenres = []
for ca in cate:
    if "Dogs" not in ca and "Rape" not in ca:
        cat = re.split(r"[->-]", ca)
        if "&" == cat[0][0]:
            cat[0] = cat[0].replace("& ", "")
        genre = cat[0].strip()
        genres.append(genre)
        if len(cat) > 2 and cat[1] != "":
            subgenres.append([genre, cat[1].strip()])

gens = np.unique(np.array(genres))
ges = sorted([g.replace('"', "") for g in gens.tolist()])
subges = sorted([[ges.index(sg[0].replace('"', "")), sg[1]] for sg in subgenres],key=lambda l:l[1])
unique_subgenres = []
subgenres_set = set()
# print(subges)

for item in subges:
    if item[1] not in subgenres_set:
        subgenres_set.add(item[1])
        unique_subgenres.append(item)
    else:
        pass
    
index_subges = ([sg[1] for sg in unique_subgenres])
print(index_subges)
new_genres = [g for g in gens if g not in index_subges]
genres = [[i + 1, new_genres[i].replace('"','')] for i in range(len(new_genres))]
subgenres = [[i + 1, unique_subgenres[i][0], unique_subgenres[i][1]] for i in range(len(unique_subgenres))]
```

```python
books_cate = df_book[["id", "category"]]
books_gen = []
for bc in books_cate.values:
    g_list = [
        ct.strip()
        for ct in bc[1]
        .replace("[", "")
        .replace("]", "")
        .replace("'", "")
        .replace("&amp;", "&")
        .replace("&gt;", ">")
        .split(",")
    ]
    book_gen = []
    for g in g_list:
        if "Dogs" not in g and "Rape" not in g:
            cat = re.split(r"[->-]", g)
            if "&" == cat[0][0]:
                cat[0] = cat[0].replace("& ", "")
            genre = cat[0].strip()
            book_gen.append(genre)
            if len(cat) > 2 and cat[1] != "":
                book_gen.append(cat[1].strip())
    books_gen.append([bc[0] + 1, book_gen])

books_gen
```

```python
books_categories = []
for bg in books_gen:
    books_genres = []
    books_subgenres = []
    for b in bg[1]:
        gen_index = -1
        subgenres_index = -1
        try:
            subgenres_index = index_subges.index(b.replace('"', ""))
        except:
            gen_index = ges.index(b.replace('"', ""))
        finally:
            books_genres.append(gen_index + 1)
            books_subgenres.append(subgenres_index + 1)
    books_categories.append(
        [
            bg[0],
            np.unique(np.array([bg for bg in books_genres if bg != 0])).tolist(),
            np.unique(np.array([bsg for bsg in books_subgenres if bsg != 0])).tolist(),
        ]
    )

books_categories
```

```python
book_genres = []
book_subgenres = []
for bc in books_categories:
    # count = 0
    # for bg in bc[1]:
    #     if bg < 1056 and count < 1:
    #         book_genres.extend([[bc[0], bg]])
    #         count += 1
    #     else:
    #         break
    book_genres.extend([[bc[0], bg] for bg in bc[1] if bg < 1056])
    book_subgenres.extend([[bc[0], bg] for bg in bc[2]])
print(book_genres)
print(book_subgenres)
```
