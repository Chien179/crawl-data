# Import Library

```python
import pandas as pd
```

# Read CSV

```python
reviews = pd.read_csv("data/reviews.csv", encoding='utf-8')
reviews.columns
```

```python
books = pd.read_csv("data/books_db.csv", encoding='utf-8')
books.columns
```

```python
users = pd.read_csv("data/users.csv", encoding='utf-8')
user_ids = [i for i in range(1,len(users.values)+1)]
data = {"user_id":user_ids, "username":users['username'].tolist()}
user_rcm = pd.DataFrame(data)
user_rcm
```

# Data

```python
datas = []
for review in reviews.values:
    book = books[books['id'] == review[2]]
    user_id = user_rcm[user_rcm['username'] == review[1]]['user_id'].values[0]
    data = [user_id, 
            review[-2], 
            book['book_title'].values[0], 
            book['author'].values[0], 
            book['pushlisher'].values[0], 
            book['images'].values[0],
            book['desc'].values[0],
            book['category'].values[0]]
    datas.append(data)
df = pd.DataFrame(datas, columns=['user_id', 
                                  'rating', 
                                  'book_title', 
                                  'author', 
                                  'publisher',
                                  'imgs',
                                  'desc', 
                                  'category'])
df
```

```python
df.to_csv("data_train/preprocessed_data.csv", sep=',', index=False, encoding='utf-8', index_label=None)
```
