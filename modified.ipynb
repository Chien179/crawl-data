# Import Library

```python
import numpy as np
import pandas as pd
import csv
from datetime import datetime
```

# Functions

```python
def has_numbers(input: str):
    return any(char.isdigit() for char in input)
```

```python
# -*- coding: utf-8 -*-
def is_english(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True
```

```python
def save_file(name_file,header, data):
    with open(name_file+'.csv', 'w') as file:
        # 2. Create a CSV writer
        writer = csv.writer(file)
        # 3. Write data to the file
        writer.writerow(header)
        writer.writerows(data)
```

# Read CSV

```python
df = pd.read_csv("data_raw/books_raw.csv", encoding='utf-8')
df
```

# Modified Data

## Repare Data Modified

```python
data_modified = df.copy()
data_modified.reindex
data_modified
```

### Categorise

```python
list_cate = []
mapping_table = str.maketrans({"'": '', 
                               '[': '', 
                               ']': ''})

for i in range(len(data_modified['category'])):
    cate = data_modified['category'][i].translate(mapping_table).replace('&amp;', '&').split(",")
    cate = [cat.strip() for cat in cate if not has_numbers(cat) and cat != " Awards"]
    if len(cate) == 1:
        cate[0] = "Orther Books"
    list_cate.append(cate)
data_modified['category'].update(pd.Series(list_cate))
data_modified['category']
```

### Price

```python
data_modified["price"] = [p.replace("&nbsp;","") for p in data_modified["price"]]
data_modified['price']
```

### Images

```python
list_images = []
mapping_table = str.maketrans({"'": '', 
                               '[': '', 
                               ']': ''})
for i in range(len(data_modified['images'])):
    images = data_modified['images'][i].split("', '")
    images = [image.translate(mapping_table).strip() for image in images]
    list_images.append(images)
data_modified['images'].update(pd.Series(list_images))
data_modified["images"]
```

### Ratings

```python
mapping_table = str.maketrans({'(': '', 
                               ')': ''})
data_modified["rating"] = [r.translate(mapping_table) for r in data_modified["rating"]]
data_modified['rating']
```

### Reviews

```python
mapping_table = str.maketrans({"'": '',
                               '"': '', 
                               '[': '', 
                               ']': ''})
list_reviews = []
for i in range(len(data_modified['reviews'])):
    reviews = data_modified['reviews'][i].split("], [")
    lst_revs = []
    for j in range(len(reviews)):
        rev = reviews[j].translate(mapping_table).split(",")
        rev = [r.strip() for r in rev if type(r) != int]
        if len(rev) == 1:
            rev[0] = None
        rev.insert(0,i)
        lst_revs.append(rev)
    list_reviews.append(lst_revs)
data_modified["reviews"] = list_reviews
data_modified["reviews"]
```

### Save

```python
data_modified.to_csv("data_raw/books_modified.csv", sep=',', index=True, encoding='utf-8', index_label="id")
```

## Repare Data Reviews DB

```python
reviews_db = data_modified['reviews'].copy()
reviews_db
```

```python
list_rev = []
reviews_table = pd.DataFrame(columns=['books_id',
            'liked',
            'disliked',
            'reported',
            'comments',
            'rating',
            'created_at'])
for rev in reviews_db:
    for r in rev:
        if len(rev) != 1:
            temp = {
                'books_id': r[0],
                'liked': r[3],
                'disliked': r[4],
                'reported': False,
                'comments':r[2],
                'is_deleted':False,
                'rating':r[1],
                'created_at': datetime.strptime(r[-1].replace("-","/"), '%Y/%m/%d')
                }
            if has_numbers(r[3]):
                reviews_table.loc[len(reviews_table)] = temp
reviews_table
```

```python
reviews_table.to_csv("data_raw/reviews_db.csv", sep=',', index=True, encoding='utf-8', index_label="id")
```
