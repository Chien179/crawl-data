# Import Library

```python
import numpy as np
import pandas as pd
import csv
from datetime import datetime
from random import randint, choice
import bcrypt
```

# Functions

```python
def get_username(name):
    return name+str(randint(100, 100000))
```

```python
def get_phone_number() -> str:
    phone_header = ['86','96','97','98','32','33','34','35','36','37','38','39','88','91','94','83','84','85','81','82','89','90','93','70','79','77','76','78','92','56','58','99','59']
    return '+84'+str(choice(phone_header))+str(randint(1000000,9999999))
```

```python
def get_age():
    return randint(12,60)
```

```python
def get_email(name):
    return name+str(randint(000, 999))+'@gmail.com'
```

```python
def hash_password():
    # example password 
    password = '123456789'
    
    # converting password to array of bytes 
    bytes = password.encode('utf-8') 
    
    # generating the salt 
    salt = bcrypt.gensalt() 
    
    # Hashing the password 
    hash = bcrypt.hashpw(bytes, salt) 
    
    return hash
```

# Read CSV

```python
df_male = pd.read_csv("data_raw/male.csv", encoding='utf-8')
len_rows = df_male.shape[0]
df_male['username'] = [get_username(i) for i in df_male['full_name']]
df_male['email'] = [get_email(i) for i in df_male['full_name']]
df_male['password'] = [hash_password() for _ in range(len_rows)]
df_male['phone_number'] = [get_phone_number() for _ in range(len_rows)]
df_male['age'] = [get_age() for _ in range(len_rows)]
df_male['sex'] = [1]*len_rows
df_male['image'] = ['https://res.cloudinary.com/doqhasjec/image/upload/v1681990980/samples/NMC%20Bookstore/Default_ct9xzk.png?fbclid=IwAR1-YjqZGj-wZzoRzdaUFg581OdLLb-3Epm7vlJF_vnCty5K-UJre4iYTpY'] * len_rows
df_male
```

```python
df_female = pd.read_csv("data_raw/female.csv", encoding='utf-8')
len_rows = df_female.shape[0]
df_female['username'] = [get_username(i) for i in df_female['full_name']]
df_female['email'] = [get_email(i) for i in df_female['full_name']]
df_female['password'] = [hash_password() for _ in range(len_rows)]
df_female['phone_number'] = [get_phone_number() for _ in range(len_rows)]
df_female['age'] = [get_age() for _ in range(len_rows)]
df_female['sex'] = [0]*len_rows
df_female['image'] = ['https://res.cloudinary.com/doqhasjec/image/upload/v1681990980/samples/NMC%20Bookstore/Default_ct9xzk.png?fbclid=IwAR1-YjqZGj-wZzoRzdaUFg581OdLLb-3Epm7vlJF_vnCty5K-UJre4iYTpY'] * len_rows
df_female
```

# Data

```python
columns = ['username','full_name', 'email', 'password', 'image', 'phone_number', 'age', 'sex']
df_male = df_male[columns]
df_female = df_female[columns]
df = pd.concat([df_male, df_female])
df = df.sample(frac=1).reset_index(drop=True)
df
```

```python
df.to_csv("data_raw/users.csv", sep=',', index=False, encoding='utf-8', index_label=None)
```
